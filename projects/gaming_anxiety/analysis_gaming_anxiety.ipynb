## Transform some data to simplify dataset

```python
# Create some new variables
data_simplified = raw_data.copy()

# condense the "earnings" column 
earnings_cat_counts = data_simplified["earnings"].value_counts()
# display(earnings_cat_counts)
earnings_low_vol_cats = earnings_cat_counts[earnings_cat_counts <= 3].index
data_simplified['earnings_condensed'] = data_simplified['earnings']
data_simplified['earnings_condensed'] = data_simplified['earnings_condensed'].replace(earnings_low_vol_cats, 'other')
# display(data_simplified['earnings_condensed'].value_counts())

# condense the "whyplay" column 
whyplay_cat_counts = data_simplified["whyplay"].value_counts()
# display(whyplay_cat_counts)
whyplay_low_vol_cats = whyplay_cat_counts[whyplay_cat_counts <= 26].index
data_simplified['whyplay_condensed'] = data_simplified['whyplay']
data_simplified['whyplay_condensed'] = data_simplified['whyplay_condensed'].replace(whyplay_low_vol_cats, 'other')
# display(data_simplified['whyplay_condensed'].value_counts())

# condense the "Playstyle" column 
playstyle_cat_counts = data_simplified["Playstyle"].value_counts()
# display(playstyle_cat_counts)
playstyle_low_vol_cats = playstyle_cat_counts[playstyle_cat_counts < 48].index
data_simplified['Playstyle_condensed'] = data_simplified['Playstyle']
data_simplified['Playstyle_condensed'] = data_simplified['Playstyle_condensed'].replace(playstyle_low_vol_cats, 'other')
# display(data_simplified['Playstyle_condensed'].value_counts())

# condense the "Residence" column 
residence_cat_counts = data_simplified["Residence"].value_counts()
# display(residence_cat_counts)
residence_low_vol_cats = residence_cat_counts[residence_cat_counts < 100].index
data_simplified['Residence_condensed'] = data_simplified['Residence']
data_simplified['Residence_condensed'] = data_simplified['Residence_condensed'].replace(residence_low_vol_cats, 'other')
# display(data_simplified['Residence_condensed'].value_counts())

# normalize and combine total psych questionnaire values
data_simplified["GAD_T_norm"] = (data_simplified["GAD_T"] - 0) / (21)  # MinMax scaled given range of 0-21
data_simplified["SPIN_T_norm"] = (data_simplified["SPIN_T"] - 0) / (68)  # MinMax scaled given range of 0-68
data_simplified["SWL_T_norm"] = (data_simplified["SWL_T"] - 5) / (30) # MinMax scaled given range of 5-35

data_simplified["SWL_T_norm_neg"] = (1 - data_simplified["SWL_T_norm"])

# average of normalized total scores (could be weighted...)
data_simplified["Psych_Total"] = (data_simplified["GAD_T_norm"] + data_simplified["GAD_T_norm"] + data_simplified["SWL_T_norm_neg"]) / 3
```

## Create some lists of constants for later use

```python
# Constants
numerical_predictors = ["Narcissism", "Hours", "Age", "streams"]
categorical_predictors = ["Game", "Platform", "earnings_condensed", "whyplay_condensed", "Gender", "Degree", "Residence_condensed", "Playstyle_condensed",]
psych_totals = ["GAD_T", "SWL_T", "SPIN_T"]
psych_totals_norm = ["GAD_T_norm", "SWL_T_norm_neg", "SPIN_T_norm", "Psych_Total"]
```
